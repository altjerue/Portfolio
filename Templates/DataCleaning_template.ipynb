- **Profiling Data**
- **Missing Values**

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
```

## Load the dataset

```python
df = pd.read_csv('your_dataset.csv')
```

## Profiling Data

```python
# Display the first few rows of the dataset
df.head()

# Display basic information about the dataset
df.info()

# Display summary statistics of the dataset
df.describe()
```

## Missing values

```python
# Check for missing values
df.isnull().sum()

# Check null % per column
df.isnull().mean().sort_values(ascending=False)

# Drop any row with null
df.dropna()

# Drop only if those columns are null
df.dropna(subset=['col1', 'col2'])

# Visualize missing values
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
```

## Duplicates

```python
# Check for fully duplicated rows
df.duplicated().sum()

# Check for duplicates across subset of columns
column_names = ['first_column', 'second_column']
df.duplicated(subset=column_names, keep=False)

# Removing duplicates
df.drop_duplicates(inplace=True)
```

## Imputation

```python
# Fill null with zero
df['col'].fillna(0)

# Fill null with forward fill
df['col'].fillna(method='ffill')

# Fill with median (numerical)
df['col_num'].fillna(df['col_num'].median(), inplace=True)

# Fill with mode (categorical)
df['col_cat'].fillna(df['col_cat'].mode()[0], inplace=True)
```

## Outliers

```python
# Detect IQR
q1 = df['col'].quantile(0.25)
q3 = df['col'].quantile(0.75)
iqr = q3 - q1

# Detect outliers
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr
outliers = df[(df['col'] < lower) | (df['col'] > upper)]

# Remove outliers
df_no_outliers = df[~outliers]
```
